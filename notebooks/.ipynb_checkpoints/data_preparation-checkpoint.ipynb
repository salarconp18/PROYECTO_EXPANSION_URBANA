# EVALUACIÓN DE APTITUD TERRITORIAL PARA IDENTIFICACIÓN DE ZONAS DE EXPANSIÓN URBANA PARA MUNICIPIOS INTERMEDIOS

## OBJETIVO


Identificar áreas potenciales de expansión urbana considerando criterios demográficos, ambientales, físicos y de accesibilidad, a través de herramientas de análisis espacial.

### PREPARACIÓN DE DATOS 

```python
import geopandas as gpd
import rasterio as rs
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
```

### Datos Usados 

1. **Limite Municipal** – Shapefile del limite de Interes - IGAC 
2. **Limites centros poblados** Shapefile de los limites de los centros poblados -DANE
3. **Clasificación uso de suelo** – Shapefile Clasificación uso de suelo - POT  
4. **Coberturas de la tierra** – Shapefile Mapa de Cobertura de la Tierra. Adaptación Corine Land Cover. Escala 1:100.000. Periodo 2018- IDEAM
5. **Zonificación amenaza por remoción en masa** - Shapefile Zonificación de Amenazas por movimientos en masa - SGC
6. **Vías** - shapefile de vias y limite vias, omitiendo caminos
7. **Drenajes sencillos** -shapefile de drenajes intermitentes y permanestes - MINTIC
8. **Canales sencillos** - shapefile de canales sencillos - MINTIC
9. **Drenajes Dobles** - shapefile de drenajes dobles - MINTIC
10. **Lagunas**- shapefile de cuerpos de agua - MINTIC
11. **Modelo Digital de terreno** - Modelo digital de terreno del Municipio, resolución 1 m,Año 2023 - IGAC

```python
Ruta_Carpeta=r"/notebooks/PROYECTO_FINAL/Datos"

Municipio_interes = Ruta_Carpeta+'/TOCANCIPA.shp'
Centros_poblados = Ruta_Carpeta+'/Centros_Poblados_Cundinamarca.shp'
Clasificacion_suelo = Ruta_Carpeta+'/Clasificacion_suelo_Cundinamarca.shp'
Coberturas_tierra = Ruta_Carpeta+'/Coberturas_Cundinamarca.shp'
Zon_Amenazas = Ruta_Carpeta+'/Remocion_Cundinamarca.shp'
drenaje_sencillo = Ruta_Carpeta+'/drenaje_sencillo.shp'
canal_sencillo = Ruta_Carpeta+'/canal_sencillo.shp'
drenaje_doble = Ruta_Carpeta+'/drenaje_doble.shp'
lagunas = Ruta_Carpeta+'/laguna.shp'
vias_vias = Ruta_Carpeta+'/via.shp'
vias_lvias = Ruta_Carpeta+'/limite_via.shp'
DTM = Ruta_Carpeta+'/MDT_Tocancipa.tif'
```

```python
# Cargue de datos vectoriales
municipio = gpd.read_file(Municipio_interes).to_crs(epsg=9377)
Centros_pob = gpd.read_file(Centros_poblados).to_crs(epsg=9377)
uso_suelo = gpd.read_file(Clasificacion_suelo).to_crs(epsg=9377)
coberturas = gpd.read_file(Coberturas_tierra).to_crs(epsg=9377)
Amenaza = gpd.read_file(Zon_Amenazas).to_crs(epsg=9377)
drenaje_sencillo = gpd.read_file(drenaje_sencillo).to_crs(epsg=9377)
canal_sencillo = gpd.read_file(canal_sencillo).to_crs(epsg=9377)
drenaje_doble = gpd.read_file(drenaje_doble).to_crs(epsg=9377)
lagunas = gpd.read_file(lagunas).to_crs(epsg=9377)
vias_vias = gpd.read_file(vias_vias).to_crs(epsg=9377)
vias_lvias = gpd.read_file(vias_lvias).to_crs(epsg=9377)
```

```python
# Cargue de datos raster
with rs.open(DTM) as src:
    dtm = src.read(1)  
    transform = src.transform
    crs = src.crs
    pixel_size_x = transform[0]
    pixel_size_y = -transform[4]  
```

```python
# Funcion de recorte 
import sys
sys.path.append('../src')  # subir un nivel y entrar a src
from data_preprocessing import clip_vector  # importar la función
```

### Pre-procesamiento Datos

```python
# Filtrado de capas

# Amenaza de movimientos en masa alta y muy alta
Amenaza_alta = Amenaza[Amenaza['Amenaza'].isin(['4. Alta', '5. Muy Alta'])]

# Coberturas bosques
Bosques = coberturas[coberturas['nivel_1'] == '3. Bosques y áreas seminaturales']

# Coberturas agrícolas
coberturas_agricolas = coberturas[coberturas['nivel_1'] == '2. Territorios agrícolas']

Suelo_Proteccion = uso_suelo[uso_suelo['Tipo_Suelo'] == 'Suelo de protección']

Zonas_Expansion_Existente = uso_suelo[uso_suelo['CS_TIPO'] == 'Expansión urbana']
```

```python
# Mapa de pendientes a partir del MDT 

# Calculo derivadas espaciales (gradientes)
gy, gx = np.gradient(dtm, pixel_size_y, pixel_size_x)

# Calcular pendiente en porcentaje
slope_percent = np.sqrt(gx**2 + gy**2) * 100

# Visualización mapa de pendientes
plt.figure(figsize=(10, 6))
plt.imshow(slope_percent, cmap='terrain', vmin=0, vmax=100)
plt.colorbar(label='Pendiente (%)')
plt.title("Mapa de Pendiente (%)")
plt.axis('off')
plt.show()
```

```python
# area del municipio 
area_municipio = municipio.geometry.area.iloc[0]
# Suma total de longitud de vías
longitud_total_vias = vias_vias['SHAPE_Leng'].sum()
longitud_total_lvias = vias_lvias['SHAPE_Leng'].sum()
# Densidad vial
longitud_total = longitud_total_vias+longitud_total_lvias
densidad_vial = (longitud_total / area_municipio) * 1e6  # m/km²
print(f"Densidad vial de Tocancipa: {densidad_vial:.2f} m/km²")
```

```python
from data_preprocessing import buffer_areas_influencia
```

```python
# Generación areas de influencia (buffer) 
# Rondas hídricas
# Diccionario para aplicar la funcion de bff
capas_hidrografia = {
    "drenaje_sencillo": drenaje_sencillo,
    "canal_sencillo": canal_sencillo,
    "drenaje_doble": drenaje_doble,
    "lagunas": lagunas
}
# diccionarios 
buffers_hidrografia = {}

for nombre, capa_hidrografia in capas_hidrografia.items():
    gdf_buffer = buffer_areas_influencia(capa_hidrografia, distancia_buffer=30, epsg_proyeccion=9377)
    buffers_hidrografia[nombre + "_buffer"] = gdf_buffer
    
    # Mostrar cada uno en su propia figura
    fig, ax = plt.subplots(figsize=(6, 6))
    gdf_buffer.plot(ax=ax)
    ax.set_title(f"Buffer de {nombre}")
    ax.axis('off')
    plt.show()
```

```python
#Buffer cercania vias
# Diccionario para aplicar la funcion de bff
capas_vias = {
    "vias_vias": vias_vias,
    "vias_lvias": vias_lvias,
}
# diccionarios de los buffer
buffers_vias = {}

for nombre, capa_vias in capas_vias.items():
    gdf_buffer = buffer_areas_influencia(capa_vias, distancia_buffer=1000, epsg_proyeccion=9377)
    buffers_vias[nombre + "_buffer"] = gdf_buffer
    
    # Crear figura para cada par capa+buffer
    fig, ax = plt.subplots(figsize=(6, 6))
    # Dibujar el buffer con transparencia
    gdf_buffer.plot(ax=ax, color='lightblue', edgecolor='blue', alpha=0.5)

    # Dibujar la capa de vías encima del buffer
    capa_vias.plot(ax=ax, color='black', linewidth=1)
    municipio.plot(ax=ax, color='none', edgecolor='black', linewidth=1.5)
    # Ajustes del gráfico
    ax.set_title(f"Buffer de {nombre}")
    ax.axis('off')
    plt.show()
```

```python
#Cercania centros poblados

gdf_buffer = buffer_areas_influencia(centros_pob_clip, distancia_buffer=1000, epsg_proyeccion=9377)

fig, ax = plt.subplots(figsize=(6, 6))
municipio.plot(ax=ax, color='white', edgecolor='black')

gdf_buffer.plot(ax=ax, color='lightblue', edgecolor='blue', alpha=0.5)

centros_pob_clip.to_crs(epsg=9377).plot(ax=ax, color='red', markersize=10)

leyenda = [
    Patch(facecolor='lightblue', edgecolor='blue', label='Buffer 1 km'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=8, label='Centros poblados')
]

ax.set_title("Buffer de proximidad a centros poblados (1 km)")
ax.axis('off')
ax.legend(handles=leyenda)

plt.show()
```
