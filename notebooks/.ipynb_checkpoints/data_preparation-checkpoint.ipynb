### 2.1 Cargue de los datos

Lo primero que se realiza es el cargue de los datos para el an√°lisis espacial y evaluaci√≥n multicriterio. Para garantizar la correcta superposici√≥n e interoperabilidad entre los datos, todos son reproyectados al sistema de referencia EPSG:9377 (Magna_sirgas_Origen_Nacional), que es adecuado para an√°lisis espaciales precisos a nivel de Colombia.

```python
Ruta_Carpeta=r"/notebooks/PROYECTO_FINAL/Datos"

Municipio_interes = Ruta_Carpeta+'/TOCANCIPA.shp'
Centros_poblados = Ruta_Carpeta+'/Centros_Poblados_Cundinamarca.shp'
Clasificacion_suelo = Ruta_Carpeta+'/Clasificacion_suelo_Cundinamarca.shp'
Coberturas_tierra = Ruta_Carpeta+'/Coberturas_Cundinamarca.shp'
Zon_Amenazas = Ruta_Carpeta+'/Remocion_Cundinamarca.shp'
drenaje_sencillo = Ruta_Carpeta+'/drenaje_sencillo.shp'
canal_sencillo = Ruta_Carpeta+'/canal_sencillo.shp'
drenaje_doble = Ruta_Carpeta+'/drenaje_doble.shp'
lagunas = Ruta_Carpeta+'/laguna.shp'
vias_vias = Ruta_Carpeta+'/via.shp'
vias_lvias = Ruta_Carpeta+'/limite_via.shp'
DTM = Ruta_Carpeta+'/MDT_Tocancipa.tif'
```

```python
# Cargue de datos vectoriales
municipio = gpd.read_file(Municipio_interes).to_crs(epsg=9377)
Centros_pob = gpd.read_file(Centros_poblados).to_crs(epsg=9377)
uso_suelo = gpd.read_file(Clasificacion_suelo).to_crs(epsg=9377)
coberturas = gpd.read_file(Coberturas_tierra).to_crs(epsg=9377)
Amenaza = gpd.read_file(Zon_Amenazas).to_crs(epsg=9377)
drenaje_sencillo = gpd.read_file(drenaje_sencillo).to_crs(epsg=9377)
canal_sencillo = gpd.read_file(canal_sencillo).to_crs(epsg=9377)
drenaje_doble = gpd.read_file(drenaje_doble).to_crs(epsg=9377)
lagunas = gpd.read_file(lagunas).to_crs(epsg=9377)
vias_vias = gpd.read_file(vias_vias).to_crs(epsg=9377)
vias_lvias = gpd.read_file(vias_lvias).to_crs(epsg=9377)
```

```python
# Cargue de datos raster
with rs.open(DTM) as src:
    dtm = src.read(1)  
    transform = src.transform
    crs = src.crs
    pixel_size_x = transform[0]
    pixel_size_y = -transform[4]  
```

### 3.1 Filtrado de capas 

Antes de llevar a cabo el an√°lisis multicriterio, se realiza un proceso de filtrado tem√°tico de las capas geogr√°ficas con el fin de extraer √∫nicamente los subconjuntos de informaci√≥n que son pertinentes para el objetivo del estudio. Este paso permite reducir la complejidad de los datos y enfocar el an√°lisis en las variables para delimitar zonas de expansi√≥n urbana.

Los filtros aplicados se enfocan en las siguientes categor√≠as:

* Zonas con amenaza alta y muy alta de remoci√≥n en masa, extra√≠das de la capa de amenazas, para identificar √°reas con restricciones por riesgo geot√©cnico.

* Coberturas de tipo "bosques y √°reas seminaturales", seleccionadas desde la capa de coberturas de la tierra, para considerar las restricciones ambientales asociadas a ecosistemas estrat√©gicos.

* Coberturas agr√≠colas, tambi√©n filtradas desde la capa de coberturas de la tierra suelo, que representan √°reas productivas relevantes para el an√°lisis de compatibilidad de usos.

* Suelos de protecci√≥n, extra√≠dos de la capa de uso actual del suelo, los cuales deben ser preservados por su valor ecol√≥gico, ambiental o de riesgo.

* Zonas de expansi√≥n urbana existentes, identificadas a partir de la clasificaci√≥n del uso actual del suelo, que indican √°reas ya previstas para el crecimiento urbano.

```python
# Amenaza de movimientos en masa alta y muy alta
Amenaza_alta = Amenaza[Amenaza['Amenaza'].isin(['4. Alta', '5. Muy Alta'])]

# Coberturas bosques
Bosques = coberturas[coberturas['nivel_1'] == '3. Bosques y √°reas seminaturales']

# Coberturas agr√≠colas
coberturas_agricolas = coberturas[coberturas['nivel_1'] == '2. Territorios agr√≠colas']

#Suelo de protecci√≥n
Suelo_Proteccion = uso_suelo[uso_suelo['Tipo_Suelo'] == 'Suelo de protecci√≥n']

#Zonas de expansion actuales
Zonas_Expansion_Existente = uso_suelo[uso_suelo['CS_TIPO'] == 'Expansi√≥n urbana']
```

### 3.2 Generaci√≥n mapa de pendientes
El an√°lisis de la pendiente del terreno es un criterio fundamental en la planificaci√≥n territorial y en la identificaci√≥n de zonas aptas para la expansi√≥n urbana. Generar un mapa de pendientes permite evaluar la topograf√≠a del √°rea de estudio, identificando las zonas con mayor facilidad para urbanizar y aquellas donde existen restricciones f√≠sicas para el desarrollo.

```python
# Mapa de pendientes a partir del MDT 

# Calculo derivadas espaciales (gradientes)
gy, gx = np.gradient(dtm, pixel_size_y, pixel_size_x)

# Calcular pendiente en porcentaje
slope_percent = np.sqrt(gx**2 + gy**2) * 100

# Visualizaci√≥n mapa de pendientes
plt.figure(figsize=(10, 6))
plt.imshow(slope_percent, cmap='terrain', vmin=0, vmax=100)
plt.colorbar(label='Pendiente (%)')
plt.title("Mapa de Pendiente (%)")
plt.axis('off')
plt.show()
```

### 3.3 C√°lculo de la densidad vial del municipio

La densidad vial es un indicador que permite evaluar el grado de infraestructura vial presente en un territorio. Se define como la longitud total de v√≠as por unidad de superficie, y se expresa com√∫nmente en metros por kil√≥metro cuadrado (m/km¬≤). Este indicador es √∫til para analizar la accesibilidad, conectividad urbana y cobertura del sistema vial, aspectos claves en la identificaci√≥n de zonas aptas para la expansi√≥n urbana.

Para representar espacialmente la variabilidad de este indicador dentro del municipio, se emplea un enfoque basado en grillas (cuadr√≠culas). Este m√©todo consiste en superponer una malla regular sobre el territorio, calcular la longitud total de v√≠as contenida en cada celda, y luego dividirla por el √°rea de la celda para obtener la densidad vial local.

```python
# area del municipio 
area_municipio = municipio.geometry.area.iloc[0]
# Suma total de longitud de v√≠as
longitud_total_vias = vias_vias['SHAPE_Leng'].sum()
longitud_total_lvias = vias_lvias['SHAPE_Leng'].sum()
# Densidad vial
longitud_total = longitud_total_vias+longitud_total_lvias
densidad_vial = (longitud_total / area_municipio) * 1e6  # m/km¬≤
print(f"Densidad vial de Tocancipa: {densidad_vial:.2f} m/km¬≤")
```

### 3.4 √Åreas influencia - rondas hidricas

Para la identificaci√≥n de zonas de expansi√≥n urbana, es fundamental delimitar las rondas h√≠dricas, entendidas como las franjas de protecci√≥n ambiental que rodean cuerpos y cursos de agua, tales como drenajes, canales y lagunas. La legislaci√≥n ambiental colombiana proh√≠be el desarrollo urbano dentro de estas zonas de ronda, con el fin de preservar la integridad de las fuentes h√≠dricas y evitar afectaciones ecol√≥gicas. Por esta raz√≥n, es indispensable su exclusi√≥n dentro del an√°lisis espacial de expansi√≥n urbana. Por tanto, las rondas h√≠dricas se delimitan mediante la generaci√≥n de buffers (√°reas de influencia) alrededor de elementos h√≠dricos

```python
from data_preprocessing import buffer_areas_influencia
```

```python
# Generaci√≥n areas de influencia (buffer) 
# Rondas h√≠dricas
# Diccionario para aplicar la funcion de bff
capas_hidrografia = {
    "drenaje_sencillo": drenaje_sencillo,
    "canal_sencillo": canal_sencillo,
    "drenaje_doble": drenaje_doble,
    "lagunas": lagunas
}
# diccionarios 
buffers_hidrografia = {}

for nombre, capa_hidrografia in capas_hidrografia.items():
    gdf_buffer = buffer_areas_influencia(capa_hidrografia, distancia_buffer=30, epsg_proyeccion=9377)
    buffers_hidrografia[nombre + "_buffer"] = gdf_buffer
    
    # Mostrar cada uno en su propia figura
    fig, ax = plt.subplots(figsize=(6, 6))
    gdf_buffer.plot(ax=ax)
    ax.set_title(f"Buffer de {nombre}")
    ax.axis('off')
    plt.show()
```

### 3.5 √Åreas influencia - cercan√≠a a v√≠as

En la identificaci√≥n de zonas de expansi√≥n urbana, es clave considerar el nivel de accesibilidad del territorio. Una de las formas m√°s efectivas de evaluar esto es mediante la generaci√≥n de √°reas de influencia (buffers) alrededor de las v√≠as existentes.

```python
#Buffer cercania vias
# Diccionario para aplicar la funcion de bff
capas_vias = {
    "vias_vias": vias_vias,
    "vias_lvias": vias_lvias,
}
# diccionarios de los buffer
buffers_vias = {}

for nombre, capa_vias in capas_vias.items():
    gdf_buffer = buffer_areas_influencia(capa_vias, distancia_buffer=1000, epsg_proyeccion=9377)
    buffers_vias[nombre + "_buffer"] = gdf_buffer
    
    # Crear figura para cada par capa+buffer
    fig, ax = plt.subplots(figsize=(6, 6))
    # Dibujar el buffer con transparencia
    gdf_buffer.plot(ax=ax, color='lightblue', edgecolor='blue', alpha=0.5)

    # Dibujar la capa de v√≠as encima del buffer
    capa_vias.plot(ax=ax, color='black', linewidth=1)
    municipio.plot(ax=ax, color='none', edgecolor='black', linewidth=1.5)
    # Ajustes del gr√°fico
    ax.set_title(f"Buffer de {nombre}")
    ax.axis('off')
    plt.show()
```
